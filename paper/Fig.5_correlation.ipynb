```python
ebs = pd.read_csv(rootpath+'numax/sample/samples_eb.csv')

vrard = pd.read_csv('~/Onedrive/Work/database/catalog/vrard+16.csv')

ebs.merge(vrard[['KIC', 'DPi1', 'e_DPi1']], on='KIC', how='left')
```

```python
from sklearn.neighbors import KernelDensity
from scipy.stats import pearsonr


# kic = 10001167 
# kic = 4663623
# kic = 5640750
# kic = 7037405
kic = 9540226

# kic = 9970396


# kic = 9153621

for kic in [kic]: #ebs['KIC'].unique():
    star = ebs[ebs['KIC']==kic].iloc[0]

    filepath = rootpath+'numax/results_gaia_nu_sc/{:0.0f}/data.h5'.format(kic)
    if not os.path.exists(filepath): continue
    f = h5py.File(filepath, 'r')

    chi2 = f['chi2_seismic'][:]
    idx = chi2 < 25

    cols = ['Dnu_freq', 'FeH', 'Teff', 'Yinit', 'Zinit', 'amlt', 
            'chi2_seismic', 'chi2_seismic_l0', 'chi2_seismic_l2', 
            'density', 'eps', 'fov_core', 'fov_shell', 'index', 'luminosity', 
            'numax_scaling', 'profile_number', 'radius', 'star_age', 'star_mass']
    df = pd.DataFrame({key: f[key][idx] for key in cols})
    df['numax'] = df['star_mass'] * df['radius']**-2.0 * (df['Teff']/5772)**-0.5 * 3090.

    df['chi2_lum'] = ((df['luminosity']-star['luminosity'])/star['e_luminosity'])**2.
    df['chi2_mass'] = ((df['star_mass']-star['star_mass'])/star['e_star_mass'])**2.
    df['chi2_radius'] = ((df['radius']-star['radius'])/star['e_radius'])**2.
    df['chi2_numax'] = ((df['numax']-star['numax'])/star['e_numax'])**2.
    df['chi2_Teff'] = ((df['Teff']-star['Teff'])/star['e_Teff'])**2.
    df['chi2_feh'] = ((df['FeH']-star['[M/H]'])/star['e_[M/H]'])**2.

    df['chi2_1'] = df['chi2_seismic'] + df['chi2_feh'] + df['chi2_Teff'] + df['chi2_lum'] 
    df['chi2_2'] = df['chi2_seismic'] + df['chi2_feh'] + df['chi2_Teff'] + df['chi2_lum']  + df['chi2_numax'] #+ df['chi2_Teff'] + df['chi2_feh']
    df['chi2_3'] = df['chi2_seismic'] + df['chi2_feh'] + df['chi2_Teff'] + df['chi2_lum']  + df['chi2_radius'] + df['chi2_mass'] #+ df['chi2_feh'] + df['chi2_Teff'] 
    df['chi2_4'] = df['chi2_seismic'] + df['chi2_feh'] + df['chi2_Teff'] + df['chi2_lum']  + df['chi2_lum'] + df['chi2_mass'] + df['chi2_radius']+ df['chi2_numax']


    fig, axes = plt.subplots(figsize=figsize(column="double", ratio=0.6), nrows=2, ncols=3, squeeze=False, constrained_layout=True)
    # fig, axes = plt.subplots(figsize=figsize(column="double", ratio=0.9), nrows=3, ncols=3, squeeze=False, constrained_layout=True)
    # axes = axes.reshape(-1)

    # each row - vs diff quantities
    xrows = ['radius', 'numax', ] #, 'Yinit'] #, 'Teff']
    xrows_latex = [r'Radius (R$_\odot$)', r'$\nu_{\rm max}$ ($\mu$Hz)', r'$Y_{\rm init}$']
    labels = [r'Common constraints: $\chi^2_{\rm freq}$, $\chi^2_{\rm spec}$, $\chi^2_{\rm Gaia}$',
              r'Common constraints: $\chi^2_{\rm freq}$, $\chi^2_{\rm spec}$, $\chi^2_{\rm Gaia}$' + '\n' + r'Additional constraints: $\chi^2_{\nu_{\rm max}}$',
              r'Common constraints: $\chi^2_{\rm freq}$, $\chi^2_{\rm spec}$, $\chi^2_{\rm Gaia}$' + '\n' + r'Additional constraints: $\chi^2_{\rm dyn}$', ]

    for irow, xrow in enumerate(xrows):
        # each col - diff chi2
        for icol in range(3):

            # rescale data for Gaussian kernel estimations
            X = df[[xrow, 'amlt']].to_numpy() # np.array([numax_single, amlt_single]).T
            scaledX = (X-X.min(axis=0))/(X.max(axis=0)-X.min(axis=0))
            kde = KernelDensity(kernel='gaussian', bandwidth=0.1).fit(scaledX, sample_weight=np.exp(-df[f'chi2_{icol+1:0.0f}']/2))
            log_density = kde.score_samples(scaledX)
            density = np.exp(log_density)

            samples = kde.sample(n_samples=10000, random_state=0)
            coeff = pearsonr(samples[:,0], samples[:,1])[0]

            # create a grid for contour
            gx, gy = np.linspace(X[:,0].min(), X[:,0].max(), 10), np.linspace(X[:,1].min(), X[:,1].max(), 10)
            g = np.array([[igx, igy ] for igy in gy for igx in gx])
            scaledgz = (g-X.min(axis=0))/(X.max(axis=0)-X.min(axis=0))
            gz = np.exp(kde.score_samples(scaledgz).reshape((len(gy), len(gx))))


            norm = matplotlib.colors.Normalize(vmin=0., vmax=density.max())

            axes[irow, icol].contourf(
                gx, gy, gz, 
                cmap='Purples', norm=norm, levels=20,
            )
            # axes[irow, icol].scatter(
            #     X[:,0], X[:,1], 
            #     marker='.',  c='gray', #cmap='cividis', norm=norm,
            # )

            if irow == 0:
                axes[irow, icol].axvline(star['radius'], c='k')
                axes[irow, icol].fill_betweenx(
                    [0,5],  [star['radius']-star['e_radius']]*2, 
                    x2=[star['radius']+star['e_radius']]*2,
                    color='gray', alpha=0.4, label='Observed '+r'$R$'
                )

            if irow == 1:
                axes[irow, icol].axvline(star['numax'], c='k')
                axes[irow, icol].fill_betweenx(
                    [0,5],  [star['numax']-star['e_numax']]*2, 
                    x2=[star['numax']+star['e_numax']]*2,
                    color='gray', alpha=0.4, label='Observed '+r'$\nu_{\rm max}$'
                )


            # if irow == 2:
            #     axes[irow, icol].axvline(star['Teff'], c='k')
            #     axes[irow, icol].fill_betweenx(
            #         [0,5],  [star['Teff']-star['e_Teff']]*2, 
            #         x2=[star['Teff']+star['e_Teff']]*2,
            #         color='gray', alpha=0.4, label='Observed '+r'$T_{\rm eff}$'
            #     )
            
            axes[irow, icol].set_xlim(X[:,0].min(), X[:,0].max())
            axes[irow, icol].set_ylim(X[:,1].min(), X[:,1].max())
            axes[irow, icol].set_xlabel(xrows_latex[irow])
            axes[irow, icol].set_ylabel(r'$\alpha_{\rm MLT}$')
            axes[irow, icol].text(0.96, 0.96, 
                                  labels[icol] + '\n\n' + f'$r = {coeff:0.2f}$', 
                                  transform=axes[irow, icol].transAxes, ha='right', va='top', color='black',
                                  fontsize=fontsize-2)


filename = 'Fig.4_correlation.pdf'
plt.savefig(filename, dpi=300)
os.system(to_overleaf(filename, 'figures'))
```
